```python
def volume_in_µl_to_pipette(V_total_ml: float, c_final: float, c_stock: float):
    V_F_buffer_ml = 0.1 * V_total_ml # 10% of total
    V_F_buffer_µL = V_F_buffer_ml * 1000
    print(f"Pipette {V_F_buffer_µL} µL des F-buffer")  

    V_G_actin_ml = V_total_ml * c_final/c_stock # c_stock [µmol/l]
    V_G_actin_µL = V_G_actin_ml * 1000
    print(f"Pipette {V_G_actin_µL} µL des actin")

    V_G_buffer_ml = V_total_ml - V_G_actin_ml - V_F_buffer_ml
    V_G_buffer_µL = V_G_buffer_ml * 1000
    print(f"Pipette {V_G_buffer_µL} µL des F-buffer")  

    test_sum = V_F_buffer_ml + V_G_actin_ml + V_G_buffer_ml
    print(test_sum)
    if test_sum == V_total_ml:
        return print("GoGo")
```

```python
def linear_fit(area, pressure, xmin, xmax):
    """Linear fit + R² for a selected interval."""
    mask = (area >= xmin) & (area <= xmax)
    x = area[mask]
    y = pressure[mask]

    if len(x) < 2:
        raise ValueError(f"Not enough points in interval {xmin}–{xmax}")

    # Linear regression
    slope, intercept = np.polyfit(x, y, 1)
    y_fit = slope * x + intercept

    # Compute R²
    ss_res = np.sum((y - y_fit)**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    r2 = 1 - ss_res/ss_tot

    return slope, intercept, x, y_fit, r2
```

```python
import numpy as np

# ============================================================
# 1) SLOPE-WERTE AUTOMATISCH AUS DEINEN FIT-BLOCKS SAMMELN
# >> WICHTIG <<  NICHTS eintragen — s_LE1, b_LE1 usw. müssen
# bereits im Notebook berechnet worden sein.
# ============================================================
slopes = {
    1: {"T": 293.55, "s_LE": s_LE1, "b_LE": b_LE1, "s_LC": s_LC1, "b_LC": b_LC1},
    2: {"T": 294.35, "s_LE": s_LE2, "b_LE": b_LE2, "s_LC": s_LC2, "b_LC": b_LC2},
    3: {"T": 296.00, "s_LE": s_LE3, "b_LE": b_LE3, "s_LC": s_LC3, "b_LC": b_LC3},
}
# Falls du statt K lieber °C in der Regression willst, setze hier:
# slopes[1]["T"] = 20.4  # statt 293.55
# slopes[2]["T"] = 21.3
# slopes[3]["T"] = 23.0

# ============================================================
# 2) Übergangsdruck π_M berechnen aus Schnittpunkt LE/LC
# ============================================================
def pi_M_from_intersection(s_LE, b_LE, s_LC, b_LC):
    """Berechnet π_M (mN/m) am Schnittpunkt der beiden Geraden."""
    if abs(s_LE - s_LC) < 1e-12:
        raise ValueError("LE- und LC-Geraden sind nahezu parallel.")
    A_M = (b_LC - b_LE) / (s_LE - s_LC)   # Fläche (Å²) am Schnittpunkt
    return s_LE * A_M + b_LE              # π_M in mN/m

for k, d in slopes.items():
    d["pi_M"] = pi_M_from_intersection(d["s_LE"], d["b_LE"], d["s_LC"], d["b_LC"])

# ============================================================
# 3) dπ_M/dT VIA LINEARER REGRESSION
# ============================================================
def linear_regression(x, y):
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)
    if len(x) < 2:
        raise ValueError("Zu wenige Punkte für Regression")
    A = np.vstack([x, np.ones(len(x))]).T
    slope, intercept = np.linalg.lstsq(A, y, rcond=None)[0]
    y_pred = slope * x + intercept
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r2 = 1 - ss_res / ss_tot if ss_tot != 0 else np.nan
    # Standardfehler der Steigung (falls >=3 Punkte)
    if len(x) > 2:
        s_err = np.sqrt(ss_res / (len(x) - 2))
        sxx = np.sum((x - np.mean(x)) ** 2)
        sigma_slope = s_err / np.sqrt(sxx) if sxx != 0 else np.nan
    else:
        sigma_slope = np.nan
    return slope, intercept, r2, sigma_slope

T_list = [slopes[i]["T"] for i in sorted(slopes.keys())]
pi_list = [slopes[i]["pi_M"] for i in sorted(slopes.keys())]

d_piM_dT, intercept, r2_pi, sigma_d_piM_dT = linear_regression(T_list, pi_list)

# ============================================================
# 4) Δq_M BERECHNEN — TRAGE HIER DEINE ΔA-WERTE EIN (Å²)
# ============================================================
delta_A = {1: 58.511, 2: 57.491, 3: 48.504}   # Beispielwerte in Å²
conv = 1e-20  # Å² -> m²

for k, d in slopes.items():
    A_m2 = delta_A[k] * conv
    T = d["T"]              # falls T in K gespeichert ist, bleibt so
    # dπM/dT in mN/(m*K) -> *1e-3 -> N/(m*K)
    d["dq_M"] = T * A_m2 * d_piM_dT * 1e-3   # J / Molekül (Annäherung)

# ============================================================
# 5) AUSGABE
# ============================================================
print("\n===== Ergebnisse =====\n")
for k in sorted(slopes.keys()):
    d = slopes[k]
    print(f"Messung {k}:")
    print(f" T = {d['T']:.2f}")
    print(f" s_LE = {d['s_LE']:.6f}")
    print(f" s_LC = {d['s_LC']:.6f}")
    print(f" π_M = {d['pi_M']:.4f} mN/m")
    print(f" Δq_M = {d['dq_M']:.3e} J/Molekül\n")

print("===== Gemeinsame Regression =====")
print(f"dπ_M/dT = {d_piM_dT:.6e} mN/(m*K)")
print(f"Intercept = {intercept:.4f} mN/m, R² = {r2_pi:.5f}, σ_slope = {sigma_d_piM_dT}")
```
